##Introduction

In recent few years, air traffic number is escalating, according to International Air Transport Association (IATA) forecast, the number of air travellers will reach 7.2 billion in 2035 (Sankararaman, et.al 2017). The increase in air traffic rate is alarming; hence, the air transportation system is focusing on system safety metrics to avoid hazardous incidents.  To facilitate the paradigm of proactive safety, a hybrid predictive model is developed for examination past incidences reports, study associated risk to predict possible occurrences using a probabilistic decision tree. 
The incident reports were accessed from the database of the Aviation Safety Reporting System; the heterogeneity of data was reduced by dividing the dataset into structured and unstructured dataset. As a result, Support Vector Machine (SVM) model could identify the relationship between event summary and risk factor of incident consequences (unstructured data), parallel to this Deep Neural Network (DNN) model was trained with the association between incident features and outcomes (structured data).
The hybrid model (Woźniak 2014) is a four-step process, consisting of the SVM model and Ensemble DNN model to predict event risk as well as event level outcome. The effectiveness and performance of the hybrid model were measured by comparing it against three different individual models, statistical tests and a ten-fold cross-validation process. The objective of the developed hybrid model was achieved by the prevailing individual model's accuracy features like recall, precision and F1 score (Zhang & Mahadevan 2019).

Apart from investigating accidents induced by human factors, the system reliability can be improved by reinforcing the operator's capabilities and efficiency with better decision making in response to hazardous events, thus mitigating the severity of accidents. In that context, unlike other methodologies, the proposed method aligns with second strategy improving aviation and system reliability by appropriate prediction of consequences of abnormal events w.r.t. Risk levels. This goal was accomplished in four steps: 
1.	Categorizing all events based on risk level into groups – low, moderately medium, medium, moderately high, high which are linked with event consequences. 
2.	Developing the SVM model to discover an association between event summary (textual form) and event consequence. Along with this, an ensemble of trained DNN establishes a complicated association linking incident features and outcomes. 
3.	Developing unique fusion rule to combine predictions resulted from two different trained ML models, therefore raising prediction quality.
4.	At last, a probabilistic decision tree is used to encompass event-level probabilities along with categorized risks level records.
The accident reports were delivered by pilots, technicians, aviation controllers and other authoritative crew to Aviation Safety Reporting System (ASRS), it was a challenge to create a model from ASRS information, the challenges faced were: 
1.	High-dimensional data: Every report consisted of 50 plus attributes like weather, flight phase, flight conditions, visibility, etc., making it hard for ML algorithms to gain accurate prediction. 
2.	Overall Categorical data: An attribute (crew size) was integer, 99% of data were categorical, as categorical features are more descriptive and less informative it was hard to blend the outcomes and maintain model performance.
3.	Unstructured data: Mining casual relationships from textual data like event synopsis, was performed by converting it into numerical data, further serving its term frequency to the SVM model.  
4.	Imbalance class distribution: Figure below represents the records (possible outcomes) are highly imbalanced across different classes. In the data reported by ASRS, records in one class were significantly larger than other.

####2.	Model Construction 

#####2.1	SVM
SVM provides high prediction accuracy; it classifies text and transforms text data into numerical attribute vectors for text representation. It discovered the relation among event summary and risk associated with incident outcome by working on unstructured data. The TF-IDF vector (Term Frequency – Inverse Document Frequency) is fed to the SVM model to recognize the hyperplane which distinguishes various classes (Lilleberg, et.al 2015). 

#####2.2	The ensemble of Deep Neural Networks (DNN)
The prime advantage of DNN is that it can extract attributes hidden in raw data with automatic representation at multiple levels (Evermann, et.al 2017). The attributed with empty records were removed, leaving behind 29 different items in a structural format. The potent ability of DNN could identify hidden layers with 24 and 12 neurons.



####3.	Model Fusion

The innovative fusion rule is established to integrate the predictions resulted from 2 different trained ML models shown in figure 5. This fusion strategy could resolve the disagreed predictions as no model is perfect; the valuable information from misclassification and erroneous predictions were used to correct model prediction. In consideration with three factors like Record-level prediction probabilities, Relation of each class record with dissimilar predictions and Class-level accuracy, the fusion rule could successfully assign test record with the highest probability (likelihood) to class as an output (Zhang & Mahadevan 2019).
